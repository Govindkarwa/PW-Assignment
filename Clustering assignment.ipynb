#### **Question 1:  What is the difference between K-Means and Hierarchical Clustering? Provide a use case for each.** 

K-Means and Hierarchical Clustering are two widely used clustering techniques in unsupervised machine learning. Both have distinct methodologies, advantages, and limitations, making them suitable for different scenarios.

**K-Means Clustering**

K-Means is a centroid-based algorithm that partitions data into a predefined number of clusters (K). It minimizes the sum of squared distances between data points and their respective cluster centroids. The algorithm iteratively assigns points to the nearest centroid and updates the centroids until convergence.

**Use Cases:**

- Large datasets: K-Means is computationally efficient and scales well with large datasets.

- Known number of clusters: It is ideal when the number of clusters (K) is predefined or can be estimated using methods like the Elbow Method.

- Compact clusters: Works best when clusters are spherical or compact in shape.

- Customer Segmentation in Marketing – A retail company can group customers into k segments (e.g., budget shoppers, premium buyers, deal seekers) based on purchase behavior.This helps in targeted advertising.

**Hierarchical Clustering**

Hierarchical Clustering builds a hierarchy of clusters represented as a dendrogram. It can be agglomerative (bottom-up) or divisive (top-down). Unlike K-Means, it does not require the number of clusters to be specified beforehand.

**Use Cases:**

- Exploratory data analysis: Ideal when the number of clusters is unknown, as the dendrogram helps determine the optimal number of clusters.

- Small to medium-sized datasets: Suitable for datasets where computational complexity is manageable.

- Hierarchical relationships: Useful when a nested structure of clusters is needed, such as in biological data analysis.

- Document Clustering in Text Mining – Useful in organizing research papers or news articles into a hierarchy of topics (e.g., “Science → Physics → Quantum Physics”).Researchers can cut the dendrogram at a desired level to get clusters.

##### Key Differences

- **Cluster Formation:** K-Means uses centroids, while Hierarchical Clustering builds a tree-like structure.

- **Number of Clusters:** K-Means requires a predefined K, whereas Hierarchical Clustering determines clusters dynamically.

- **Scalability:** K-Means is faster and better suited for large datasets, while Hierarchical Clustering is more computationally intensive.

- **Visualization:** Hierarchical Clustering provides a dendrogram, while K-Means does not offer a hierarchical view.

#### **Question 2: Explain the purpose of the Silhouette Score in evaluating clustering algorithms.**

When we use clustering algorithms like K-Means to group data, we need a way to check how good those groups are. The Silhouette Score is one of the most popular ways to do this. It helps us measure how well each data point fits into its assigned cluster and how far it is from other clusters.

The Silhouette Score is a metric used to evaluate the quality of clustering results. It measures how well-defined and separated the clusters are by calculating a value for each data point based on two factors:

Cohesion: How similar a data point is to other data points within its own cluster. This is measured by the average distance from a point to all other points in the same cluster.

Separation: How dissimilar a data point is to points in other clusters. This is measured by the average distance from a point to all points in the nearest neighboring cluster.

The score for a single data point ranges from -1 to +1, where:

- A score close to +1 indicates that the data point is well-matched to its own cluster and well-separated from neighboring clusters. This is the ideal result.

- A score close to 0 suggests that the data point is near the decision boundary between two clusters, meaning the clusters are overlapping.

- A score close to -1 indicates that the data point may have been assigned to the wrong cluster.

To get an overall score for the entire clustering, you take the average of the Silhouette Scores for all data points. This average score provides a quantitative measure of the clustering's quality, which is especially useful in unsupervised learning tasks where there are no true labels to validate the clusters. A higher average Silhouette Score indicates better, more consistent clustering. This metric is also often used to help determine the optimal number of clusters for a dataset.

##### Purpose in Evaluation

- Helps compare different clustering algorithms (e.g., K-Means vs Hierarchical).

- Helps choose the right number of clusters (k) in methods like K-Means.

- Provides an objective metric instead of relying only on visual inspection.



##### **Question 3: What are the core parameters of DBSCAN, and how do they influence the clustering process?** 

DBSCAN, short for Density-Based Spatial Clustering of Applications with Noise, is a powerful clustering algorithm used in machine learning. Unlike traditional clustering methods like K-Means, DBSCAN identifies clusters based on the density of data points, making it particularly effective for datasets with arbitrary shapes and noise.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a clustering algorithm that groups together points that are close to each other based on density, while marking outliers as noise.

It has two core parameters that strongly influence the clustering process:

1. eps (Neighborhood Radius): This defines the radius of the neighborhood around a data point. If the distance between two points is less than or equal to eps they are considered neighbors. A common method to determine eps is by analyzing the k-distance graph.\

Influence:
- A small ε → leads to many small clusters (or most points may be labeled as noise if too small).
- A large ε → merges more points into the same cluster, possibly combining distinct clusters.

2. MinPts (Minimum Points): This is the minimum number of points required within the eps radius to form a dense region. A general rule of thumb is to set MinPts >= D+1 where D is the number of dimensions in the dataset.

Influence:
- A small MinPts (e.g., 2 or 3) → even small dense regions form clusters, but more noise may be incorrectly clustered.
- A large MinPts → requires denser regions to form clusters, which can eliminate weak/noisy clusters but may also mark too many points as noise.
- For most cases a minimum value of MinPts = 3 is recommended.

**How DBSCAN Works**

- Identify core points by counting neighbors within the ε radius.

- Form clusters by expanding from core points and connecting density-reachable points.

- Label points that do not belong to any cluster as noise.

This process allows DBSCAN to handle clusters of arbitrary shapes and effectively manage noise.


#### **Question 4: Why is feature scaling important when applying clustering algorithms like K-Means and DBSCAN?**

Feature scaling is a critical preprocessing step for clustering algorithms like K-Means and DBSCAN because these algorithms rely on distance-based metrics to determine similarity between data points. Without scaling, features with a larger magnitude or wider range of values can disproportionately influence the distance calculation, leading to biased and inaccurate clustering results.

1. Distance Measures are Sensitive to Scale

- K-Means uses Euclidean distance to assign points to clusters.

- DBSCAN also relies on distance (ε-neighborhood) to decide which points are “close” enough to form clusters.

- If features are on very different scales (e.g., age in years = 0–100, income in thousands = 0–100,000), the feature with the larger scale will dominate the distance computation, leading to biased clusters.

2. Fair Contribution of Features

- Scaling ensures that all features contribute equally to the clustering process.

- Without scaling, one feature with a large numeric range can overshadow others, even if the smaller-scale features carry meaningful patterns.

3. Effect on K-Means

- Centroids are computed as the mean of feature values.

- If features aren’t scaled, centroids get pulled toward large-scale features, producing distorted clusters.

- Example: In customer segmentation, “income” (0–1,00,000) will dominate over “age” (0–100) if not scaled.

4. Effect on DBSCAN

- DBSCAN groups points based on ε-radius.

- If features are unscaled, ε will work well for one dimension but not for another, leading to poor cluster detection (either merging too many points or labeling too many as noise).


#### **Question 5: What is the Elbow Method in K-Means clustering and how does it help determine the optimal number of clusters?**

The Elbow Method is a popular heuristic used to determine the optimal number of clusters (k) for K-Means clustering. It relies on the concept of Within-Cluster Sum of Squares (WCSS), which measures the sum of the squared distances between each data point and the centroid of its assigned cluster. Essentially, WCSS quantifies how tightly grouped the clusters are. A lower WCSS value indicates better, more compact clusters.

The Elbow Method is a simple but effective technique used to determine the optimal number of clusters (K) in a K-Means clustering algorithm. It examines the relationship between the number of clusters and the within-cluster sum of squares (WCSS), a measure of the variance within each cluster.

**How It Works**
The Elbow Method involves the following steps:

**1. Run K-Means for a range of k values:** You start by running the K-Means algorithm for a variety of possible cluster numbers, for example, from k=1 to k=10.

**2. Calculate WCSS for each k:** For each value of k, you calculate the total WCSS. As you increase the number of clusters, the WCSS will naturally decrease because data points will be closer to their respective centroids. For k=1, the WCSS is at its maximum, as all data points are in a single cluster.


**3. Plot the results:** You then plot the WCSS values on the y-axis against the number of clusters (k) on the x-axis.

**4. Identify the "Elbow":** The resulting graph will typically show a steep drop in WCSS at first, as the first few clusters significantly improve the model's ability to represent the data. After a certain point, the rate of decrease in WCSS slows down and becomes more linear. This point of inflection, where the curve bends like an elbow, is considered the optimal number of clusters. Beyond this point, adding more clusters provides only marginal improvement, which could lead to overfitting.

#### **Question 6: Generate synthetic data using make_blobs(n_samples=300, centers=4),apply KMeans clustering, and visualize the results with cluster centers.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Step 2: Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Step 3: Visualize the clustered data and cluster centers
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Data Points')

# Plot the cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Cluster Centers')

plt.title("K-Means Clustering with 4 Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data using make_blobs
X, y_true = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_pred = kmeans.fit_predict(X)

# Get cluster centers
centers = kmeans.cluster_centers_

# Create visualization
plt.figure(figsize=(12, 5))

# Plot: KMeans clustering results with centers
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', alpha=0.7,label='Data Points')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, linewidths=3, label='Centroids')
plt.title('KMeans Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.tight_layout()
plt.show()

# Print results
print("KMeans Clustering Results:")
print(f"Number of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print(f"Number of clusters: {kmeans.n_clusters}")
print(f"Inertia (within-cluster sum of squares): {kmeans.inertia_:.2f}")

print("\nCluster Centers:")
for i, center in enumerate(centers):
    print(f"Cluster {i}: ({center[0]:.2f}, {center[1]:.2f})")

# Count points in each cluster
unique, counts = np.unique(y_pred, return_counts=True)
print("\nPoints per cluster:")
for cluster, count in zip(unique, counts):
    print(f"Cluster {cluster}: {count} points")
```

#### **Question 7: Load the Wine dataset, apply StandardScaler , and then train a DBSCAN model. Print the number of clusters found (excluding noise).**

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Step 1: Load the Wine dataset
data = load_wine()
X = data.data

# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=5)  # You can tune these parameters
dbscan.fit(X_scaled)

# Step 4: Count number of clusters (excluding noise)
labels = dbscan.labels_

# Noise points are labeled as -1
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

print(f"Number of clusters found (excluding noise): {n_clusters}")
```

#### **Question 8: Generate moon-shaped synthetic data using make_moons(n_samples=200, noise=0.1), apply DBSCAN, and highlight the outliers in the plot.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate moon-shaped synthetic data
X, y_true = make_moons(n_samples=200, noise=0.1, random_state=42)

# Scale the data (recommended for DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
cluster_labels = dbscan.fit_predict(X_scaled)

# Count clusters and outliers
unique_labels = np.unique(cluster_labels)
n_clusters = len(unique_labels[unique_labels != -1])
n_outliers = np.sum(cluster_labels == -1)

# Create visualization
plt.figure(figsize=(15, 7))


# Plot 1: DBSCAN clustering results
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', alpha=0.7)
plt.title('DBSCAN Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot 2: Highlighting outliers
plt.subplot(1, 2, 2)
# Plot normal points
normal_points = cluster_labels != -1
outliers = cluster_labels == -1

plt.scatter(X[normal_points, 0], X[normal_points, 1], 
           c=cluster_labels[normal_points], cmap='viridis', alpha=0.7, label='Clusters')
plt.scatter(X[outliers, 0], X[outliers, 1], 
           c='red', marker='x', s=100, label='Outliers')
plt.title('DBSCAN with Highlighted Outliers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.tight_layout()
plt.show()

# Print results
print("DBSCAN on Moon Data Results:")
print(f"Number of samples: {X.shape[0]}")
print(f"Number of clusters found: {n_clusters}")
print(f"Number of outliers: {n_outliers}")

print(f"\nDBSCAN Parameters:")
print(f"eps: {dbscan.eps}")
print(f"min_samples: {dbscan.min_samples}")

# Calculate outlier percentage
outlier_percentage = (n_outliers / len(X)) * 100
print(f"Outlier percentage: {outlier_percentage:.1f}%")
```

#### **Question 9: Load the Wine dataset, reduce it to 2D using PCA, then apply Agglomerative Clustering and visualize the result in 2D with a scatter plot.**

```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load Wine dataset
X, y = load_wine(return_X_y=True)

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Apply Agglomerative Clustering
clustering = AgglomerativeClustering(n_clusters=3)
labels = clustering.fit_predict(X_pca)

# Step 4: Visualize results
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='Set1', s=50)
plt.title('Agglomerative Clustering on Wine Dataset (2D PCA)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()

print(f"Original shape: {X.shape}")
print(f"PCA shape: {X_pca.shape}")
print(f"Number of clusters: {clustering.n_clusters}")
```
