**Question 1:  What is Simple Linear Regression?**

Simple linear regression is a statistical and supervised learning method in which a single independent variable (also known as a predictor variable) is used to predict the dependent variable. In other words, it models the linear relationship between the dependent variable and a single independent variable.

**1. Independent Variable (Predictor, X):**

 The feature inputs in the dataset are termed as the independent variables. There is only a single independent variable in simple linear regression. An independent variable is also known as a predictor variable as it is used to predict the target value. It is plotted on a horizontal axis.

**2. Dependent Variable (Response, Y):**

 The target value in the dataset is termed as the dependent variable. It is also known as a response variable or predicted variable. It is plotted on a vertical axis.

Simple linear regression helps make predictions and understand relationships between one independent variable and one dependent variable. For example, you might want to know how a tree’s height (independent variable) affects the number of leaves it has (dependent variable). By collecting data and fitting a simple linear regression model, you could predict the number of leaves based on the tree's height. This is the ‘making predictions’ part. But this approach also reveals how much the number of leaves changes, on average, as the tree grows taller, which is how simple linear regression is also used to understand relationships.It assumes a linear relationship between X and Y, meaning changes in X result in proportional changes in Y.



**Question 2: What are the key assumptions of Simple Linear Regression?**

Simple linear regression is a parametric test, meaning that it makes certain assumptions about the data. These assumptions are:

**1. Linearity −** This assumption assumes that the relationship between the dependent and independent variables is linear. That means the dependent variable changes linearly as the independent variable changes. A scatter plot will show the linearity in the dataset.

**2. Homoskedasticity −** For all observations, the variance of the residuals is the same. This assumption relates to the squared residuals.

**3. Independence −** The examples (observations or X and Y pairs) are independent. There is no collinearity in data so the residuals will not be correlated. To check this, we example the scatter plot of residuals vs. fits.

**4. Normality −** Model Residuals are normally distributed. Residuals are the differences between the actual and predicted values. To check for the normality, we examine the histogram of residuals. The histogram should be approximately normally distributed.


**Question 3: What is heteroscedasticity, and why is it important to address in regression models?**

Heteroscedasticity refers to the situation in regression analysis where the variability of the residuals (errors) is not constant across all levels of the independent variables. In other words, the spread of the residuals changes as the predicted values change. This can lead to inconsistent results and a violation of the assumptions of the model. It is important to address heteroscedasticity because it can affect the validity of the regression model's results, making it more likely that significant relationships between independent and dependent variables are not detected. Addressing heteroscedasticity ensures that the model's estimates are more reliable and that the model's assumptions are met, leading to more accurate predictions and insights.

**How to Fix It**
- Transform the dependent variable: Applying log, square root, or inverse transformations can stabilize variance.
- Use weighted least squares: This method assigns less weight to observations with higher variance.
- Robust standard errors: They help adjust for heteroscedasticity without changing the coefficients.




**Question 4: What is Multiple Linear Regression?**

Multiple linear regression refers to a statistical technique that is used to predict the outcome of a variable based on the value of two or more variables. It is sometimes known simply as multiple regression, and it is an extension of linear regression. The variable that we want to predict is known as the dependent variable, while the variables we use to predict the value of the dependent variable are known as independent or explanatory variables.

Multiple Linear Regression is a powerful statistical technique used to model the relationship between one dependent variable and two or more independent variables. It helps us understand how different factors contribute to a particular outcome.

**Why It’s Useful**
- Prediction: Forecast outcomes based on multiple input features.
- Explanation: Understand which variables have the strongest impact.
- Control: Account for confounding variables in observational data.
- Optimization: Aid decision-making in fields like finance, marketing, health, and engineering.



**Question 5: What is polynomial regression, and how does it differ from linear regression?**

Polynomial Linear Regression is a type of regression analysis in which the relationship between the independent variable and the dependent variable is modeled as an n-th degree polynomial function.Unlike simple linear regression, which models the relationship as a straight line, polynomial regression can capture non-linear patterns in the data by fitting a polynomial equation.

**Linear Regression:**
 Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data. 
The equation for simple linear regression is: y = a + b * x
Linear regression is ideal when the relationship between variables is linear, making it simple to interpret and computationally efficient. It is commonly used in scenarios like predicting house prices based on size or forecasting sales trends.

**Polynomial Regression:**
 Polynomial regression is an extension of linear regression that models the relationship between the dependent variable and the independent variable(s) as an n-th degree polynomial. 
The equation for polynomial regression is: y = a0 + a1 * x + a2 * x^2 + ... + an * x^n
This method is suitable for modeling non-linear relationships where data points form curves. Polynomial regression provides flexibility to capture complex patterns but is prone to overfitting, especially with higher-degree polynomials. It is often used in scenarios like predicting crop yields based on environmental factors or modeling electricity consumption relative to temperature.

**Key Differences**
- Linear regression assumes a straight-line relationship, making it simpler and less prone to overfitting. However, it may underfit data with non-linear trends. Polynomial regression, on the other hand, can model non-linear relationships effectively but requires careful tuning of the polynomial degree to avoid overfitting.


**Question 6:**  Implement a Python program to fit a Simple Linear Regression model to 
the following sample data: 
- X = [1, 2, 3, 4, 5] 
- Y = [2.1, 4.3, 6.1, 7.9, 10.2] 

Plot the regression line over the data points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Plot data points
plt.scatter(X, Y, color='blue', label='Actual data')
# Plot regression line
plt.plot(X, Y_pred, color='red', label='Regression line')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```


**Question 7:** Fit a Multiple Linear Regression model on this sample data: 

- Area = [1200, 1500, 1800, 2000] 
- Rooms = [2, 3, 3, 4] 
- Price = [250000, 300000, 320000, 370000] 

Check for multicollinearity using VIF and report the results. 

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Prepare the data
Area = [1200, 1500, 1800, 2000]
Rooms = [2, 3, 3, 4]
Price = [250000, 300000, 320000, 370000]

X_multi = pd.DataFrame({'Area': Area, 'Rooms': Rooms})
y_multi = np.array(Price)

# Fit Multiple Linear Regression
mlr_model = LinearRegression()
mlr_model.fit(X_multi, y_multi)

# Calculate VIF for each feature
X_with_const = np.column_stack([np.ones(X_multi.shape[0]), X_multi])
vif_data = pd.DataFrame()
vif_data["feature"] = ['Intercept'] + list(X_multi.columns)
vif_data["VIF"] = [variance_inflation_factor(X_with_const, i) for i in range(X_with_const.shape[1])]

print("VIF Results:")
print(vif_data)
```


**Question 8:**  Implement polynomial regression on the following data: 
- X = [1, 2, 3, 4, 5]
- Y = [2.2, 4.8, 7.5, 11.2, 14.7] 

Fit a 2nd-degree polynomial and plot the resulting curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Step 1: Input data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Step 2: Transform input data for polynomial regression (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Step 3: Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Step 4: Predict values
X_fit = np.linspace(1, 5, 100).reshape(-1, 1)
X_fit_poly = poly.transform(X_fit)
Y_pred = model.predict(X_fit_poly)

# Step 5: Plot the original data and the polynomial regression curve
plt.scatter(X, Y, color='blue', label='Original Data')
plt.plot(X_fit, Y_pred, color='red', label='Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2nd Degree Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```


**Question 9:** Create a residuals plot for a regression model trained on this data: 
- X = [10, 20, 30, 40, 50] 
- Y = [15, 35, 40, 50, 65] 

Assess heteroscedasticity by examining the spread of residuals. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit regression model
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals vs predicted values
plt.scatter(Y_pred, residuals, color='purple')
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()
```


**Question 10:** Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model. 

As a data scientist predicting house prices using features like area, rooms, and location, I must address heteroscedasticity and multicollinearity. These issues can weaken the reliability of my regression model. Here’s how I would tackle each one:

1. Handling Multicollinearity  
**Problem:** High correlation between independent variables, such as area and rooms, leads to unstable coefficient estimates.

**Steps to Address:**  
- **Check VIF (Variance Inflation Factor):**  
Identify which variables have a VIF greater than 5 or 10.  

- **Remove or Combine Variables:**  
Drop one of the correlated features, like rooms, if it adds redundant information. Alternatively, combine them into a new feature, such as area per room.  

- **Use Regularization Techniques:**  
Apply Ridge Regression (L2) or Lasso Regression (L1). These methods penalize large coefficients and lessen the impact of multicollinearity.  

Use Principal Component Analysis (PCA) to reduce dimensionality and decorrelate features while still keeping most of the variance.  

2. Handling Heteroscedasticity  
**Problem:** Non-constant variance of residuals violates a key assumption of linear regression.

**Steps to Address:**  
- **Residual Plot Check:**  
Plot residuals against predicted values to visually assess the changing spread.  

- **Log or Power Transformations:**  
Apply log(price), sqrt(area), etc., to stabilize variance. This is particularly useful if the variance increases with predicted values.  

- **Weighted Least Squares (WLS):**  
Assign weights that are inversely proportional to the variance. This gives less importance to high-variance observations.  

- **Use Robust Standard Errors:**  
This helps adjust inference, such as p-values and confidence intervals, even when heteroscedasticity is present. You can do this using statsmodels with HC1, HC3, etc.  

3. General Modeling Enhancements  
- **Add Location as Categorical Feature (One-hot encoding):**  
Properly encode locations to avoid the dummy variable trap by dropping one dummy.  

- **Use Tree-Based Models:**  
Models like Random Forest, XGBoost, or Gradient Boosting manage multicollinearity and heteroscedasticity better than linear regression.  

- **Cross-Validation:**  
Use methods like k-fold CV to ensure model stability across different data slices.
